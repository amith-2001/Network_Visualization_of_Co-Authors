```python
columns_to_replace = ['Publisher', 'Conference name', 'Conference date', 'Authors with affiliations', 'Abbreviated Source Title']
df[columns_to_replace] = df[columns_to_replace].fillna('')
df['Cited by'] = df['Cited by'].fillna(0)
```

```python
author_info_list = []  # List to store author information

for _, row in df.iterrows():
    current_paper_id = row['EID']
    author_ids = row['Author(s) ID'][:-1].split(";") if ";" in row['Author(s) ID'] else [row['Author(s) ID']]
    
    publication_details = {
        "Title": row['Title'],
        "Year": row['Year'],
        "Citations": row['Cited by'],
        "Publisher": row['Publisher'],
        "Country": row['Authors with affiliations'].split(";")[0].split(",")[-1].strip()
    }

    for author_id in author_ids:
        # Check if the author already exists in author_info_list
        author_exists = False
        for author_info in author_info_list:
            if author_info[0] == author_id:
                author_exists = True
                author_info[1]['publications'].append({current_paper_id: publication_details})
                break

        if not author_exists:
            # If the author doesn't exist, add them to author_info_list
            author_info_list.append((author_id, {'publications': [{current_paper_id: publication_details}]}))
```

```python
# Initialize an empty list to store links
links = []

# Iterate through author_info_list to find common publications and create edges
for i in range(len(author_info_list)):
    author_id_i, data_i = author_info_list[i]
    publications_i = data_i['publications']
    
    for j in range(i + 1, len(author_info_list)):
        author_id_j, data_j = author_info_list[j]
        publications_j = data_j['publications']
        
        # Check for common publications
        common_pubs = [
            pub_i for pub_i in publications_i for pub_j in publications_j
            if pub_i.keys() == pub_j.keys()
        ]
        
        # If there are common publications, add the link to the links list
        if common_pubs:
            links.append((author_id_i, author_id_j))
G.add_edges_from(links)
```
